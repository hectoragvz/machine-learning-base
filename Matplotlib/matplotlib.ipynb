## Matplotlib Example workflow

```python
#Importing matplotlib and getting it ready
%matplotlib inline
import matplotlib.pyplot as plt

# Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# Setup plot
fig, ax = plt.subplots(figsize=(10,10)) # width and height

# Plot the data
ax.plot(x, y)

# Customize the plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# Save and show
fig.savefig('./plot_imgs/sample_plot.png')
```

```python
# Make a plot from a dictionary
nut_butter_prices = {"Almond": 10, "Peanut": 8, "Cashew": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Butter prices")
ax.set(xlabel="Butters")
ax.set(ylabel="Prices ($)");
```

```python
# horizontal bar plot
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some more data for histograms
x = np.random.randn(100)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                              figsize=(10,5))

# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10, 5))

# Plot to each index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000));
```

### How to really plot? (pyplot vs matplotlib oo method)

* When plotting quick, use pyplot
* When more advanced plots, use the OO method

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c='target');
```

```python
# OO mixed method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax)

ax.set_xlim([45, 100]);
```

```python
# OO from scratch

fig, ax = plt.subplots(figsize=(10, 6))
# Plot the data
scatter = ax.scatter(x=over_50['age'], 
                     y=over_50['chol'],
                     c=over_50['target']);

# Custom
ax.set(title="Heart Disease and Chol levels",
       xlabel="Age", ylabel="Cholesterol")

ax.legend(*scatter.legend_elements(), title="Target");

# Add a line
ax.axhline(over_50['chol'].mean(), linestyle="--");
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(12, 10), sharex=True)

# Add the data
scatter = ax0.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])

#Custom
ax0.set(title="Heart Disease and Cholesterol Levels", ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50['chol'].mean(), linestyle="--")



# Add the data
scatter = ax1.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'])
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age", ylabel="Max Heart Rate")

ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(y=over_50['thalach'].mean(), linestyle="--")

# Adding overall title
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
